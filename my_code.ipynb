```python
#Importing the pandas module
import pandas as pd
#Reading the files and assigning variable names
articles = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Skills group project/Group 4/articles.APOE.csv")
authors = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Skills group project/Group 4/authors.APOE.csv")
paper_count = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Skills group project/Group 4/paper_counts.csv")
```

```python
#Creating a line plot of paper count over time
#Setting the title, range for x axis and axes labels
paper_count.plot(x="Year", y="Count", title="Interest in gene APOE over time",
                 xlim=(1799, 2023), ylabel="Number of papers published", xlabel="Year")
```

```python
#Creating a bar chart of number of papers published per author
#Importing numpy library
import numpy as np
#Identifying authors by their first and last name
#Making a column for how many times they appear in the articles csv file (i.e. number of papers published by that author)
freq_firstauthor = pd.crosstab(articles["FirstAuthorForename"], articles["FirstAuthorLastname"]).replace(0,np.nan).stack().reset_index().rename(columns={0:'Time'})
#Sorting the authors by most papers published, saving as a new variable
sorted = freq_firstauthor.sort_values('Time', ascending=False)
#Creating a dataframe containing top 40 authors with the most published papers
most_published_firstauthors = sorted.iloc[:40,:]
#Plotting bar chart with the number of papers of published per author
plt.bar(range(len(most_published_firstauthors)), most_published_firstauthors["Time"])
#Setting the bars so that the forename and last name are shown
#Rotating the names 90 degrees
#Setting font size
plt.xticks(range(len(most_published_firstauthors)), [f"{first} {last}" for first, last in zip(most_published_firstauthors["FirstAuthorForename"],
                                                                                              most_published_firstauthors["FirstAuthorLastname"])], rotation=90, fontsize=8)
#Labelling the axes with appropriate titles
plt.xlabel("Author")
plt.ylabel("Number of papers published")
#Setting a title for the plot
plt.title("40 most published authors within the field of gene APOE")
#Displaying the graph
plt.show()
```
